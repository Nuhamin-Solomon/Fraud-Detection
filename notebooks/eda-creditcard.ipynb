```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cc_df = pd.read_csv("../data/raw/creditcard.csv")
cc_df.head()
```

```python
cc_df.info()
cc_df.isnull().sum().sum()
```

```python
cc_df['Class'].value_counts(normalize=True)
sns.countplot(x='Class', data=cc_df)
plt.title("Credit Card Fraud Class Distribution")
plt.show()
```

```python
sns.boxplot(x='Class', y='Amount', data=cc_df)
plt.title("Transaction Amount vs Fraud")
plt.show()
```
