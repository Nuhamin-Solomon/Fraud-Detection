```python
import pandas as pd
import numpy as np

fraud_df = pd.read_csv("../data/raw/Fraud_Data.csv")
import os
os.makedirs("../data/processed", exist_ok=True)
import os
os.makedirs("../data/processed", exist_ok=True)

```

```python
fraud_df['signup_time'] = pd.to_datetime(fraud_df['signup_time'])
fraud_df['purchase_time'] = pd.to_datetime(fraud_df['purchase_time'])
```

```python
fraud_df['hour_of_day'] = fraud_df['purchase_time'].dt.hour
fraud_df['day_of_week'] = fraud_df['purchase_time'].dt.dayofweek

fraud_df['time_since_signup'] = (
    fraud_df['purchase_time'] - fraud_df['signup_time']
).dt.total_seconds() / 3600
```

```python
fraud_df = fraud_df.sort_values(['user_id', 'purchase_time'])

fraud_df['txn_count_24h'] = fraud_df.groupby('user_id')['purchase_time'] \
    .transform(lambda x: x.diff().dt.total_seconds().lt(86400).cumsum())
```

```python
features = [
    'purchase_value',
    'hour_of_day',
    'day_of_week',
    'time_since_signup',
    'txn_count_24h'
]

X = fraud_df[features]
y = fraud_df['class']
```

```python
processed_df = X.copy()
processed_df['class'] = y

processed_df.to_csv("../data/processed/fraud_processed.csv", index=False)

print("✅ fraud_processed.csv saved successfully")
```
